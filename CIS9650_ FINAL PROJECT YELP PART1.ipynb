```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import requests
```

```python
r=requests.get("https://www.yelp.com/biz/ricardo-steak-house-new-york")
c=r.text
print(c)
soup=BeautifulSoup(c, 'lxml')
```

```python
import re #we use regular expression here to find the total reviews in the page
num_reviews=soup.find('span' ,attrs={"class":"css-bq71j2"}).string
num_reviews=int(re.findall('\d+',num_reviews)[0]) #here we used \d+ to only take the numbers from the string 
print(num_reviews)
```
