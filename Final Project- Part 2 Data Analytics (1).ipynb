```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
```

```python
# read the csv file
data=pd.read_csv(r'cpus.csv', low_memory=False)
# print the data in brief view
print(data)
```

```python
# Examining total rows; check how many rows in the file
print('Total rows in the file: ',len(data))
# Examining total columns; check how many columns in the file
print('Total columns in the file: ', len(data.columns))
# Examining cloumn names, print columns titles in the file
print(data.columns)
```

```python
# Accessing rows(#); Print data with header
data.head(10)
```

```python
# Select specific columns from the bigger data set, remove 'unmaed:0','syct','chmin','chmax'column by not include
# Modified the table with selected range.
data2=pd.DataFrame(data, columns=['name','mmin','mmax','cach','perf','estperf'])
```

```python
# New specificed columns with selected header range.
data2.head(10)
```

```python
# group by selected column mmin/mmax to find the average cpu performance
data2.groupby(by='mmin',dropna=False).mean()
#data2.groupby(by='mmax',dropna=False).mean()
```

```python
# convert string to numeric
# Count how many times the same values appear in the content fields
data2_mmin=data['mmin'].value_counts(sort=False)
print(data2_mmin)

#data2_mmax=data['mmax'].value_counts(sort=False)
#print(data2_mmax)

#data2_name=data['name'].value_counts(sort=False)
#print(data2_name)
```

```python
#Find the top 5 cpus; get and sort largest 5 mmax values
data2.nlargest(5,['mmax'])
```

```python
#Find the bottom 5 cpus; get and sort smallest 5 mmin values
data2.nsmallest(5,['mmin'])
```

```python
# .describe() function will quickly display statistics for any available or group it applied to 
data2['mmax'].describe()
```

```python
# statistics for cpus performance
data2['perf'].describe()
```

```python
import seaborn
```

```python
# import seaborn; seaborn.distplot(data2[mmax])
# get mmax cpu perform by user perference

seaborn.displot(data2['mmax'].dropna(), kde=False)
plt.xlabel("CPU Max Performance ")
plt.ylabel("CPU Max Distribution Range")
plt.title("CPU Speed Max Performance in Distribution Range")
```

```python
seaborn.displot(data2['perf'].dropna(), kde=False)
plt.xlabel("CPU Performance Parameters")
plt.ylabel("User Group")
plt.title("CPU Performance Parameters in User Group")
```
